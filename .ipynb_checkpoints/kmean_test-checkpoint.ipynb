```python
import numpy as np
from kmeans import lloyd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.random.randn(1000, 2)
fig, ax = plt.subplots()
ax.plot(X[:, 0], X[:, 1], '.')
fig.show()
```

```python
choice_cluster, initial_state = lloyd(X, 3)

fig, ax = plt.subplots()
for i in range(3):
	indices = np.where(choice_cluster==i)[0]
	selected = X[indices]
	ax.plot(selected[:, 0], selected[:, 1], '.', label=str(i))

fig.show()
```

```python
X = [np.random.randn(1000, 2), np.random.randn(1000, 2)+5, np.random.randn(1000, 2)+10]
X = np.concatenate(X, axis=0)
fig, ax = plt.subplots()
ax.plot(X[:, 0], X[:, 1], '.')
fig.show()
```

```python
choice_cluster, initial_state = lloyd(X, 3)

fig, ax = plt.subplots()
for i in range(3):
	indices = np.where(choice_cluster==i)[0]
	selected = X[indices]
	ax.plot(selected[:, 0], selected[:, 1], '.', label=str(i))

fig.show()
```
